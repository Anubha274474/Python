```python
# Question 3:- Replace missing values in the quantity and Price columns with their respective mean values using pandas.
```

```python
# Filling Missing values in Quantity Column 
df['Quantity']=df['Quantity'].fillna(df['Quantity'].mean())
```

```python
# removing missing values from price column 
df['Price']=df['Price'].fillna(df['Price'].mean())
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
# Question 5:- Create a new column called Revenue using Numpy,where:
# Revenue = Quantity*Price
# Display the updated dataset with the new column.
```

```python
import numpy as np
```

```python
df["Revenue"]=np.multiply(df["Quantity"],df["Price"])
```

```python
df
```

```python
# Question 6:- Using Numpy,calculate the total Revenue generated from all transections in the dataset.
```

```python
total_revenue=np.sum(df["Revenue"])
print("Total Revenue from all the transections",total_revenue)
```

```python
# Question 7:-Group by Product Category-> Total Revenue per category
```

```python
Category_revenue=df.groupby("Product Category")["Revenue"].sum()
```

```python
print(Category_revenue)
```

```python
# Question 8:- Based on the revenue calculated for each product category,identify the top 3 categories with the highest
# revenue and the bottom 3 categories with the lowest revenue.
```

```python
print(Category_revenue.nlargest(3))
```

```python
print(Category_revenue.nsmallest(3))
```

```python
#Question 9:-Group the dataset by Region and Calculate the total revenue generated in each region. Identify the region 
# with the highest revenue and the region with the lowest revenue.
```

```python
Region_revenue=df.groupby("Region")["Revenue"].sum()
```

```python
print(Region_revenue)
```

```python
print("Highest revenue region wise",df.Region,Region_revenue.max())
```

```python
print("Lowest revenue region wise",df.Region,Region_revenue.min())
```

```python
# Question 10:- Extract the month from the Data column using Pandas and group the dataset by month to calculate the 
# total monthly revenue. Also,calculate the mean,median,mode and standard deviation of the Revenue column using Numpy.
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
df['Date']= pd.to_datetime(df['Date'],dayfirst=True,errors='coerce')
```

```python
print(df['Date'].head(10))
```

```python
print(df['Date'].isna().sum())
```

```python
df['Month']=df['Date'].dt.month
```

```python
Monthly_Revenue=df.groupby('Month')['Revenue'].sum()
```

```python
print(Monthly_Revenue)
```

```python
print("Mean",np.mean(df['Revenue']))
```
