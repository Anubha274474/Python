```python
# Question 3:-Customer Purchase Analysis
# Identify the top 5 customers based on total sales.
# Hint -Use GroupBy with aggregation.
```

```python
# Question 6:-Multi-Level Grouping
# Group the dataset by: Region and Category
# Calculate the total sales for each combination.
# What insight can you derive from this analysis?
```

```python
# Insights:-
```

```python
# Question 7:-Pivot Table Creation
# Create a pivot table showing:
# Rows-> Region
# Columns-> Product
# Values-> Total Sales
# Which region sells the most laptops?
```

```python
pivot=pd.pivot_table(df,index='Region',columns='Product',values='Sales',aggfunc='sum',fill_value=0)
```

```python
print('Most laptops:',pivot['Laptop'].idxmax())
```

```python
# Question 8:- Profitability Analysis
# Calculate the average profit for each product category.
# Which category is most profitable?
```

```python
df.groupby('Category')['Profit'].mean().sort_values(ascending=False)
```

```python
cat_profit=df.groupby('Category')['Profit'].mean()
```

```python
print('Category Profit:',cat_profit.idxmax())
```
