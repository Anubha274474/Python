```python
Q6- Identify anomalies where houses have high prices but low areas.
```

```python
# Anamalies - High price but Low area
```

```python
Q1_area=df['living area'].quantile(0.25)
Q3_area=df['living area'].quantile(0.75)
IQR_area=Q3_area - Q1_area
low_area_cutoff=Q1_area-1.5* IQR_area

Q3_price=df['Price'].quantile(0.75)
high_price_cutoff=Q3_price  # top 25% percent prices

anomolies=df[(df['living area']<low_area_cutoff)&(df['Price']>high_price_cutoff)]
print(anomolies[['Price','living area']])
```
