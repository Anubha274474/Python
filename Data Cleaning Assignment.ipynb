```python
# Question 6:- Outlier Detection (Treatment Cost)
scenario:
Extreame treatment costs are affecting analysis.
    Task:
* Detect outliers using IQR method
* Display number of outliers
```

```python
Q1=df['Treatment_Cost'].quantile(0.25)
Q3=df['Treatment_Cost'].quantile(0.75)
IQR=Q3-Q1
```

```python
LOWER_BOUND=Q1-1.5*IQR
UPPER_BOUND=Q3+1.5*IQR
```

```python
print(LOWER_BOUND)
print(UPPER_BOUND)
```

```python
outliers=df[(df['Treatment_Cost']<LOWER_BOUND)|(df['Treatment_Cost']>UPPER_BOUND)]
```

```python
print('Number of outliers',len(outliers))
print(outliers[['Patient_ID','Treatment_Cost']])
```

```python
# Question 7:- Outlier Treatment
Scenerio:
# The business team wants to retain all records.
Task
# Apply capping(Winsorization)on Treatment_Cost
# Use 5th and 95th percentile
```

```python
lower_cap=df['Treatment_Cost'].quantile(0.05)
upper_cap=df['Treatment_Cost'].quantile(0.95)
```

```python
df['Treatment_Cost_Capped']=df['Treatment_Cost'].clip(lower=lower_cap,upper=upper_cap)
```

```python
print('Before max',df['Treatment_Cost'].max())
print('After max',df['Treatment_Cost'].max())
```

```python
# Question 8:-Transformation
Scenario:
Treatment cost is highly skewed.
    Task:
*Apply lof transformation
*Create a new column
*Compare before vs after distribution
```

```python
df['Log_Treatment_Cost']=np.log1p(df['Treatment_Cost'])
```

```python
print('Before skew',df['Treatment_Cost'].skew())
print('After skew',df['Log_Treatment_Cost'].skew())
```
